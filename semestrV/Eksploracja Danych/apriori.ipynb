# Apriori
"Osoba która kupiła to kupi też tamto" -> Apriori analizuje dane zakupowe żeby móc wyciągnąć takie wnioski
np. osoba która kupuje płatki śniadaniowe kupi też mleko

**Definicje pojęć**

1) Support

2) Confidence
![i9878868](https://i.imgur.com/ji53xXW.png)

3) Lift
![i234](https://i.imgur.com/RLrYFwV.png)

This means that lift basically compares the improvement of an association rule against the overall dataset. If “any product => X” in 10% of the cases whereas “A => X” in 75% of the cases, the improvement would be of 75% / 10% = 7.5.

If the lift of a rule is 1, then the products are independent of each other. Any rule that has a lift of 1 can be discarded.

If the lift of a rule is higher than 1, the lift value tells you how strongly the righthand side product depends on the left-hand side.

```python
from apyori import apriori
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
MIN_SUPPORT = 0.003
MIN_CONFIDENCE = 0.2
MIN_LIFT = 3
MIN_LENGTH = 2
MAX_LENGTH = 2
```

```python
def prepare_output(results) -> pd.DataFrame:
    df = pd.DataFrame(columns=('Items','Antecedent','Consequent','Support','Confidence','Lift'))
    Support =[]
    Confidence = []
    Lift = []
    Items = []
    Antecedent = []
    Consequent=[]

    for RelationRecord in results:
        for ordered_stat in RelationRecord.ordered_statistics:
            Support.append(RelationRecord.support)
            Items.append(RelationRecord.items)
            Antecedent.append(ordered_stat.items_base)
            Consequent.append(ordered_stat.items_add)
            Confidence.append(ordered_stat.confidence)
            Lift.append(ordered_stat.lift)

    df['Items'] = list(map(set, Items))                                   
    df['Antecedent'] = list(map(set, Antecedent))
    df['Consequent'] = list(map(set, Consequent))
    df['Support'] = Support
    df['Confidence'] = Confidence
    df['Lift']= Lift
    return df
```

<H2><center>Dane są zapisane jako lista list zakupów np.</br>
<p style="color:red">[..., [burgers, meatballs, eggs], [chutney], ...]

```python
transactions = []
with open("Market_Basket_Optimisation.csv", "r") as f:
    for line in f.readlines():
        transactions.append(line.replace('\n','').split(','))

print(transactions[0])
```

```python
rules = apriori(transactions,
                min_support = MIN_SUPPORT,
                # procentowy udział transakcji zawierający lewą i prawą stronę
                # w zbiorze wszystkich tranzakcji
                min_confidence = MIN_CONFIDENCE,
                # prawdopodobieństwo warunkowe wystąpienia reguły
                # (procent transakcji zawierających X i Y w zbiorze wszystkich transakcji zawierających X)
                max_length = MAX_LENGTH,
                min_lift = MIN_LIFT
)
```

```python
results = prepare_output(rules)
```
