```python
import matplotlib.pyplot as plt
import numpy as np
import random
import matplotlib as mpl
```

```python
def z1():
    x = np.arange(0,4*np.pi,0.01)
    y = np.sin(x)
    plt.plot(x,y)
    plt.scatter(x:=np.arange(0.5*np.pi,4*np.pi,np.pi), [1 if not i % 2 else -1 for i, _ in enumerate(x)], c="red")
    plt.xlabel('x values from 0 to 4pi')
    plt.ylabel('sin(x)')
    plt.title('Plot of sin 0 to 4pi')
    plt.legend(['sin(x)', 'extremal points'])
    plt.show()
z1()
```

```python
def z2():
    no_of_balls = 100
    x = [random.triangular() for i in range(no_of_balls)]
    y = [random.gauss(0.5, 0.25) for i in range(no_of_balls)]
    colors = [random.randint(1, 4) for i in range(no_of_balls)]
    areas = [np.pi * random.randint(5, 15)**2 for i in range(no_of_balls)]
    plt.figure()
    plt.scatter(x, y, s=areas, c=colors, alpha=0.85, cmap="winter")
    plt.axis([0.0, 1.0, 0.0, 1.0])
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.show()


z2()
```

```python
def z3():
    x = np.arange(0,4*np.pi,0.01)
    fig, axs = plt.subplots(2,2)
    axs[0,0].plot(x, np.sin(x))
    axs[0,1].plot(x, np.cos(x))
    #x = np.arange(0.4*np.pi,0.6*np.pi,0.01)
    x = np. linspace(-2 * np. pi, 2 * np. pi, 1000)
    y=np.tan(x)
    y[:-1][np.diff(y) < 0] = np.nan
    axs[1,0].plot(x, y)
    axs[1,0].axis(ymin=-5,ymax=5)
    axs[1,1].axis(ymin=-5,ymax=5)
    y = np.cos(x)/np.sin(x)
    y[:-1][np.diff(y) > 0] = np.nan
    axs[1,1].plot(x, y)
    axs[0,0].set_title("sin(x)")
    axs[0,1].set_title("cos(x)")
    axs[1,0].set_title("tan(x)")
    axs[1,1].set_title("cot(x)")
    plt.show()
z3()
```

```python
x = np.arange(0,np.pi,0.1)
y = np.sin(x)
random_x = np.random.uniform(low=0, high=np.pi, size=(500,))
random_y = np.random.uniform(low=0, high=1, size=(500,))
colors = []
for value,value_y in zip(random_x,random_y):
    if value_y < np.sin(value):
        colors.append(1)
    else:
        colors.append(0)

#np.random()
plt.plot(x,y)
plt.scatter(random_x,random_y, c=colors)
plt.show()
```
